```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import os

# Set OMP_NUM_THREADS environment variable to 1
os.environ["OMP_NUM_THREADS"] = "1"

# Read the CSV file
df = pd.read_csv("college.csv")

# Display the DataFrame
print(df)

# Extract the 'Which Programming Languages are you Proficient in?' column
# Split the languages and perform one-hot encoding
languages = df['Which Programming Languages are you Proficient in?'].str.get_dummies(sep=';')

# Number of clusters
k = 3

# Apply K-means clustering
kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
predict = kmeans.fit_predict(languages)

# Print the cluster predictions
print("Cluster predictions:", predict)

# Get unique colors for each cluster using a colormap
colors = plt.cm.viridis(predict / (k - 1))

# Bar plot showing the distribution of languages
languages.sum().plot(kind='bar', color=colors, legend=False)
plt.title('Bar plot of Programming Languages Proficiency')
plt.xlabel('Programming Languages')
plt.ylabel('Number of Students Proficient')
plt.show()

# Scatter plot of clusters
plt.scatter(range(len(df)), df.index, c=colors, marker='o')
plt.title('Scatter plot of Students by Programming Language Clusters')
plt.xlabel('Student Index')
plt.ylabel('Cluster')
plt.xticks(range(len(df)), df['Name'], rotation=90)
plt.show()

# Display cluster information
df['Cluster'] = predict
print("Clusters:")
print(df[['Name', 'Cluster']])
```
