```python
# Convert data to pandas DataFrame
df = pd.DataFrame(data)
```

```python

# Function to count interests
def count_interests(df):
    interests_count = {}

    for interests in df["Primary areas of interest within computer Science field"]:
        for interest in interests.split(';'):
            interest = interest.strip()
            if interest in interests_count:
                interests_count[interest] += 1
            else:
                interests_count[interest] = 1

    return interests_count

# Count interests
interests_count = count_interests(df)
```

```python
# Extract data for plotting
interests = list(interests_count.keys())
counts = list(interests_count.values())
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.barh(interests, counts, color='skyblue')
plt.xlabel('Number of Students')
plt.ylabel('Interest')
plt.title('Number of Students Interested in Each Field')
plt.tight_layout()
plt.show()
```
