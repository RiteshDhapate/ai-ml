```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import os

# Set OMP_NUM_THREADS environment variable to 1
os.environ["OMP_NUM_THREADS"] = "1"

# Read the CSV file
df = pd.read_csv("college.csv")

# Display the DataFrame
print(df)

# One-hot encode the preferences for theoretical vs practical
preference_dummies = pd.get_dummies(df['Do you prefer courses that are more theoretical or practical in nature?'])

# Number of clusters
k = 3

# Apply K-means clustering
kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
predict = kmeans.fit_predict(preference_dummies)

# Print the cluster predictions
print("Cluster predictions:", predict)

# Add cluster labels to the original DataFrame
df['Cluster'] = predict

# Get unique colors for each cluster using a colormap
colors = plt.cm.viridis(predict / (k - 1))

# Bar plot showing the distribution of preferences
preference_dummies.sum().plot(kind='bar', color=colors, legend=False)
plt.title('Bar plot of Course Preferences (Theoretical vs Practical)')
plt.xlabel('Course Preference')
plt.ylabel('Number of Students')
plt.show()

# Count of each preference per cluster
cluster_counts = df.groupby('Cluster')['Do you prefer courses that are more theoretical or practical in nature?'].value_counts().unstack()

# Cluster distribution bar plot
cluster_counts.plot(kind='bar', stacked=True, color=colors)
plt.title('Cluster Distribution of Course Preferences')
plt.xlabel('Cluster')
plt.ylabel('Number of Students')
plt.legend(title='Course Preference')
plt.show()

# Display cluster information
print("Clusters:")
print(df[['Name', 'Cluster']])
```
