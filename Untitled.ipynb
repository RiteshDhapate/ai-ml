```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the CSV data into a DataFrame
df = pd.read_csv('college.csv')

# Drop any rows with missing values if necessary
df = df.dropna()

# Questions to be analyzed
questions = [
    'Are you currently taking any add-on or extra cocurricular courses outside of your major requirements?',
    'Do you prefer courses that are more theoretical or practical in nature?',
    'Are you interested to learn add on courses?',
    'Which Programming Languages are you Proficient in?',
    'Interested in',
    'What are your Primary areas of interest within computer Science field?',
    'Do you have any background knowledge in'
]

# Function to convert categorical data into numerical format
def preprocess_data(column):
    if column.name in ['Which Programming Languages are you Proficient in?', 'Interested in', 'What are your Primary areas of interest within computer Science field?']:
        vectorizer = CountVectorizer(tokenizer=lambda x: x.split(';'))
        X = vectorizer.fit_transform(column)
    else:
        X = pd.get_dummies(column)
    return X

# Function to plot clusters
def plot_clusters(X, title, num_clusters=3):
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X_scaled)

    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_scaled)

    plt.figure(figsize=(14, 8))
    colors = ['r', 'g', 'b']
    markers = ['o', 's', 'd']
    for cluster in range(num_clusters):
        cluster_data = X_pca[kmeans.labels_ == cluster]
        plt.scatter(cluster_data[:, 0], cluster_data[:, 1], 
                    color=colors[cluster], marker=markers[cluster], 
                    label=f'Cluster {cluster + 1}', alpha=0.5)

    plt.title(title)
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.legend()
    plt.grid(True)
    plt.show()

# Analyze each question
for question in questions:
    print(f"Analyzing: {question}")
    X = preprocess_data(df[question])
    plot_clusters(X, question)
```
